You've built a model. In this exercise you will test how good your model is.

```python
# Code you have previously used to load data
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

# Path of the file to read
iowa_file_path = '../input/home-data-for-ml-course/train.csv'

home_data = pd.read_csv(iowa_file_path)
y = home_data.SalePrice
feature_columns = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[feature_columns]

# Specify Model
iowa_model = DecisionTreeRegressor()
# Fit Model
iowa_model.fit(X, y)

print("First in-sample predictions:", iowa_model.predict(X.head()))
print("Actual target values for those homes:", y.head().tolist())

# Set up code checking
from learntools.core import binder
binder.bind(globals())
from learntools.machine_learning.ex4 import *
print("Setup Complete")
```

```python
# Let's see the mean absolute error generated by the model.

from sklearn.metrics import mean_absolute_error

iowa_model_error = mean_absolute_error(y,iowa_model.predict(X))
```

## Step 4: Calculate the Mean Absolute Error in Validation Data


```python
from sklearn.metrics import mean_absolute_error
val_mae = mean_absolute_error(val_y,val_predictions)

# uncomment following line to see the validation_mae
print(val_mae)

# Check your answer
step_4.check()
```

Remember, from our previous observation where we haven't splited the data set into training and validation set we have the MAE of only 7459.62 but after splitting the data into training and testing dataset we see the MAE has drastically increased to 29652.93, reducing the overfitting of data.
